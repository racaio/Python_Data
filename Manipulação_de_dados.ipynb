```python
import pandas as pd
```

```python
url = '/content/dados_hospedagem.json'
```

```python
dados1 =  pd.read_json(url)
dados1.head() #dados serão apresentado aninhados em dicionarios, terei de mudar isso.
```

```python
#normalização dos dados
```

```python
dados_normalizados = pd.json_normalize(dados1['info_moveis'])
dados_normalizados.head(2)
```

```python
#ajustando as colunas que estao em listas
```

```python
colunas = list(dados_normalizados.columns)
colunas # identificar colunas
```

```python
dados_explode = dados_normalizados.explode(colunas[3:]) # so preciso da posição 3 em diante /  explode ire transformar lista em linhas
dados_explode.head()
```

```python
dados_explode.reset_index(inplace=True, drop=True) # reset_index = para resgatar os index /  Drop = 'True'= para nao utiliza os indices antigos
dados_explode.head()
```

```python
dados_explode.info() # para saber os tipos de dados que possuo no data-frame
#/  Quando tipo é object isso ocorre pois mesma coluna pode ter mais de um tipo de dado(string, floac, etc) ou quando é tudo string isso acontece
```

```python
import numpy as np # para transformar o tipo de uma coluna para int necessario importar a biblioteca Numpy
```

```python
dados_explode['max_hospedes']=dados_explode['max_hospedes'].astype(np.int64) # processo para converrter uma coluna e ja adiciona-lo ao data frame
dados_explode.info()
```

```python
col_numericas = ['quantidade_banheiros','quantidade_quartos','quantidade_camas']# convertentdo mais de uma coluna
```

```python
dados_explode[col_numericas]=dados_explode[col_numericas].astype(np.int64) #
```

```python
3dados_explode.info()
```

```python
dados_explode['avaliacao_geral']=dados_explode['avaliacao_geral'].astype(np.float64) # convertendo o tipo para float
```

```python
#dados_explode.info()
```

```python
#Alterando typo de coluna de valor para float64. Antes devo tratar para retirar os caracteres
```

```python
dados_explode['preco'] = dados_explode['preco'].apply(lambda x: x.replace('$','').replace(',','').strip())
#usando uma função lambda eu faço com que o cursos rode por cada elemendo da coluna realizando as substiruiçoes de $ e ,
#stryp = remoder espaços vazios
```

```python
dados_explode['preco'] = dados_explode['preco'].astype(np.float64)
dados_explode.info()
```

```python
dados_explode.head()
```

```python
dados_explode[['taxa_deposito','taxa_limpeza']]=dados_explode[['taxa_deposito','taxa_limpeza']].applymap(lambda x: x.replace('$','').replace(',','').strip())
#para fazer com mais de uma coluna(dataframe) usar o ApplyMap, fará a mesma coisa que o Apply
```

```python
dados_explode[['taxa_deposito','taxa_limpeza']]=dados_explode[['taxa_deposito','taxa_limpeza']].astype(np.float64)
```
